```python
import pandas as pd
import numpy as np
```

Prèviament es descarreguen els datasets de https://opendata-ajuntament.barcelona.cat/data/es/organization/seguretat i es realitzen tasques de normalització de títols de columnes.

```python
all_data=pd.DataFrame()
for i in range(2010,2016):  
    yearly_data=pd.read_csv(str(i)+'_accidents_gu_bcn.csv',encoding = "ISO-8859-1",sep=';')
    all_data=pd.concat([all_data,yearly_data])
for i in range(2016,2023):
    yearly_data=pd.read_csv(str(i)+'_accidents_gu_bcn.csv',encoding = "utf_8",sep=';')
    all_data=pd.concat([all_data,yearly_data])
```

```python
all_data
```

```python
all_data=all_data.drop(columns=['Longitud','Latitud','Longitud_WGS84','Latitud_WGS84'])
all_data
```

Afegim dades de les persones involucrades en els accidents i el tipus de vehicle.

```python
persones_accidents=pd.DataFrame()
for i in range(2010,2016):  
    yearly_data=pd.read_csv('accidents_persones_gu_bcn_'+str(i)+'.csv',encoding = "ISO-8859-1",sep=';')
    persones_accidents=pd.concat([persones_accidents,yearly_data])
for i in range(2016,2023):
    yearly_data=pd.read_csv('accidents_persones_gu_bcn_'+str(i)+'.csv',encoding = "utf_8",sep=';')
    persones_accidents=pd.concat([persones_accidents,yearly_data])
```

```python
persones_accidents.columns
```

```python
persones_accidents=persones_accidents[['Numero_expedient','Desc_Tipus_vehicle_implicat','Descripcio_sexe',
                                          'Edat','Descripcio_tipus_persona','Descripcio_Lloc_atropellament_vianant',
                                         'Descripcio_Motiu_desplaçament_vianant','Descripcio_Motiu_desplaçament_conductor','Descripcio_victimitzacio']]
persones_accidents
```

S'afegeixen dades de la causa del conductor.

```python
accidents_causa_conductor=pd.DataFrame()
for i in range(2016,2022):
    yearly_data=pd.read_csv('accidents_causa_conductor_gu_bcn_'+str(i)+'.csv',encoding = "ISO-8859-1",sep=';')
    accidents_causa_conductor=pd.concat([accidents_causa_conductor,yearly_data])
for i in range(2022,2023):
        yearly_data=pd.read_csv('accidents_causa_conductor_gu_bcn_'+str(i)+'.csv',encoding = "utf_8",sep=',')
        accidents_causa_conductor=pd.concat([accidents_causa_conductor,yearly_data])
```

```python
accidents_causa_conductor.columns
```

```python
accidents_causa_conductor=accidents_causa_conductor[['Numero_expedient','Descripcio_causa_conductor']]
accidents_causa_conductor
```

I, per últim, s'afegeix si hi ha alguna causa mediata en l'accident.

```python
accidents_causes=pd.DataFrame()
for i in range(2010,2016):  
    yearly_data=pd.read_csv('accidents_causes_gu_bcn_'+str(i)+'.csv',encoding = "ISO-8859-1",sep=';')
    accidents_causes=pd.concat([accidents_causes,yearly_data])
for i in range(2016,2023):
    yearly_data=pd.read_csv('accidents_causes_gu_bcn_'+str(i)+'.csv',encoding = "utf_8",sep=';')
    accidents_causes=pd.concat([accidents_causes,yearly_data])
```

```python
accidents_causes.columns
```

```python
accidents_causes=accidents_causes[['Numero_expedient','Descripcio_causa_mediata']]
accidents_causes
```

S'observa com s'han filtrat els datasets per a facilitar el procés de join que es realitza a continuació.

```python
all_accidents_persones=all_data.merge(persones_accidents, on='Numero_expedient', how='left')
```

```python
all_accidents_persones
```

```python
all_accidents_persones_causa_mediata=all_accidents_persones.merge(accidents_causes, on='Numero_expedient', how='left')
```

```python
all_accidents_persones_causa_mediata['Nom_districte'].unique()
```

```python
all_accidents_persones_causa_mediata_conductor=all_accidents_persones_causa_mediata.merge(accidents_causa_conductor, on='Numero_expedient', how='left')
```

```python
all_accidents_persones_causa_mediata_conductor.columns
```

```python
all_accidents_persones_causa_mediata_conductor.to_csv('final_dataset.csv', sep=';',encoding='utf')
```
